# **Preprossing**

```python
# Fill missing values in categorical columns with the most frequent value (mode)

numerical_cols = df_train.select_dtypes(include=np.number).columns.tolist()

# Exclude 'id' and 'y' from numerical columns to fill if they are present and numerical
if 'id' in numerical_cols:
    numerical_cols.remove('id')
if 'y' in numerical_cols:
    numerical_cols.remove('y') # 'y' is the target, usually not imputed this way

for col in numerical_cols:
    if col in df_train.columns:
        # Ensure the column is actually numeric before calculating mean
        df_train[col] = pd.to_numeric(df_train[col], errors='coerce')
        df_train[col] = df_train[col].fillna(df_train[col].mean())

# Identify categorical columns after renaming
categorical_cols = df_train.select_dtypes(include='object').columns.tolist()

for col in categorical_cols:
    if col in df_train.columns:
        # Ensure the column is object/string before calculating mode
        df_train[col] = df_train[col].astype(str) # Ensure string type
        df_train[col] = df_train[col].fillna(df_train[col].mode()[0])
# Drop rows with NaN in the target variable 'y'
df_train_cleaned = df_train.dropna(subset=['y'])
```

```python
df_train.isna().sum()
```

```python
# Identify all categorical columns in the training dataset and apply Label Encoding to convert them into numerical values.
# This ensures that machine learning models can work with the data by replacing string categories with integer codes.

categorical_cols_train = df_train.select_dtypes(include='object').columns.tolist()

le = LabelEncoder()

for col in categorical_cols_train:
    if col in df_train.columns:
        try:
            # Convert the column to string type to handle potential mixed types
            df_train[col] = df_train[col].astype(str)
            # Apply LabelEncoder
            df_train[col] = le.fit_transform(df_train[col])
        except Exception as e:
            print(f"Could not encode column {col}: {e}")
            pass # Continue to the next column if encoding fails
```

```python
df_train.head()
```

```python
# Apply StandardScaler to all numerical columns except 'id' and target 'y'

scaler = StandardScaler()
# Identify numerical columns to scale (excluding 'id' and 'y')
numerical_cols_to_scale = df_train.select_dtypes(include=np.number).columns.tolist()

if 'id' in numerical_cols_to_scale:
    numerical_cols_to_scale.remove('id')
if 'y' in numerical_cols_to_scale:
    numerical_cols_to_scale.remove('y') # 'y' is the target, not scaled for classification


if numerical_cols_to_scale:
    df_train[numerical_cols_to_scale] = scaler.fit_transform(df_train[numerical_cols_to_scale])
    print("StandardScaler applied to numerical columns in df_train.")
else:
    print("Warning: No numerical columns identified for scaling after excluding 'id' and 'y'.")
```
